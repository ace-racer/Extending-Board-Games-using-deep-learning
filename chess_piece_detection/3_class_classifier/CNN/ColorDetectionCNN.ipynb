```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import os
import random
import keras
random.seed(42)

%matplotlib inline
```

```python
IMAGE_SIZE = (70, 70)
def process_image(image_location):
    image = cv2.imread(image_location)
    
    if image.shape[0] != IMAGE_SIZE[0] or image.shape[1] != IMAGE_SIZE[1]:
        print("Resizing the image: {0}".format(image_location))
        resized_image = cv2.resize(image, IMAGE_SIZE, interpolation = cv2.INTER_AREA)
    else:
        resized_image = image
    
    gray = cv2.cvtColor(resized_image, cv2.COLOR_BGR2GRAY)
       
    return gray
```

```python
original_image = cv2.imread("H:\\AR-ExtendingOnlineGames\\my_board\\split_images\\board1\\b_5.jpg")
plt.imshow(original_image)
original_image.shape
```

```python
gray_input_image = process_image("H:\\AR-ExtendingOnlineGames\\my_board\\split_images\\board1\\b_5.jpg")
plt.imshow(gray_input_image, cmap='gray')
```
